```python
import os
labels = pd.read_csv('train_labels.csv')

#part_2
import csv
train_labels = []
with open('train_labels.csv', newline='') as csvfile:

  spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
  for row in spamreader:
    row = row[0][0:]

    row  = row.split(',')
    
    train_labels.append(row)

print(train_labels)

#part 3
import shutil
import csv

path = 'train/train'
dir_list = os.listdir(path)
dir_list = [i.split('.tif') for i in dir_list]
dir_list = ["".join(i) for i in dir_list]


final_path = 'train'

#part_4
path = 'train/train'
```

```python
from fastai.vision.all import *
files = get_image_files(path)


def labelFunc(filename):
  filename = filename.split('.tif')
  filename = "".join(filename)
  for lst in train_labels:
    if(lst[0]==filename):
     
      label_req = lst[1]
      break

  return label_req

#Loading of the data

dls = ImageDataLoaders.from_name_func(path, files,labelFunc, item_tfms=Resize(224),batch_tfms=aug_transforms(size=224))
dls.show_batch(max_n = 16)
```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.lr_find()
```

```python
learn.fine_tune(2,3e-3)
```

```python
learn.show_results()
```

```python
interp = Interpretation.from_learner(learn)
interp.plot_top_losses(9, figsize=(15,10))
```

```python
#checking it on validation dataset
path = 'validation/validation/'
results = {'id':[],'label':[]}
for filename in os.listdir('validation/validation'):
  prediction = learn.predict(path+filename)
  filename = filename.split('.tif')
  filename = "".join(filename)
  results['id'].append(filename)
  results['label'].append(prediction[0])

new = pd.DataFrame.from_dict(results)
```
